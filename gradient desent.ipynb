```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("insurance_data.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.age,df.bought_insurance,test_size=0.2,random_state=20)
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(1,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(x_train,y_train,epochs=500)
```

```python
model.evaluate(x_test,y_test)
```
